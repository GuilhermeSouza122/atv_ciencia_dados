```python
# Puxando arquivos de √°udio 
audio_files = glob.glob("Streaming_History_Audio*.json")

dfs = []

for file in audio_files:
    with open(file, "r", encoding="utf-8") as f:
        data = json.load(f)
        df_temp = pd.DataFrame(data)
        df_temp["source_file"] = file 
        dfs.append(df_temp)


# Mesclando √°udios 
df_streaming_history = pd.concat(dfs, ignore_index=True)

# Mostrando o dataframe final de √°udio
df_streaming_history.head(5)
```

```python
df_streaming_history.info()
df_streaming_history.dropna()
```

```python
# Limpeza de dados 
df_streaming_history = df_streaming_history.drop(columns=["ip_addr"])
df_streaming_history = df_streaming_history.drop(columns=["audiobook_title", "audiobook_uri", "audiobook_chapter_uri", "audiobook_chapter_title"])
df_streaming_history = df_streaming_history.drop(columns=["source_file"]) 

# Tratamento de dados 
df_streaming_history["ts"] = pd.to_datetime(df_streaming_history["ts"])
# df_streaming_history["content_type"] = df_streaming_history.apply(
#     lambda x: "podcast" if pd.notna(x["episode_name"]) else "music",
#     axis=1
# ) # Cria√ß√£o de coluna para diferenciar m√∫sica de podcast
df_streaming_history["is_podcast"] = df_streaming_history.apply(
    lambda x: 1 if pd.notna(x["episode_name"]) else 0,
    axis=1
) # Cria√ß√£o de coluna para diferenciar m√∫sica de podcast


# Renomeando as colunas
column_rename_map = {
    "ts": "played_at",  
    "platform": "platform_used",  
    "ms_played": "milliseconds_played",  
    "conn_country": "connection_country",  

    "master_metadata_track_name": "track_name",  
    "master_metadata_album_artist_name": "artist_name",  
    "master_metadata_album_album_name": "album_name",  
    "spotify_track_uri": "track_uri",  

    "episode_name": "podcast_episode_name",   
    "episode_show_name": "podcast_show_name",  
    "spotify_episode_uri": "podcast_episode_uri",  


    "reason_start": "play_start_reason",  
    "reason_end": "play_end_reason",  

    "shuffle": "shuffle_enabled",  
    "skipped": "skipped_track",  
    "offline": "offline_playback",  
    "offline_timestamp": "offline_timestamp",  

    "incognito_mode": "incognito_mode"
}
df_streaming_history = df_streaming_history.rename(columns=column_rename_map)

# Vendo informa√ß√µes do novo dataframe
df_streaming_history.info()


```

```python
username = "Mayumi Castiglioni"

query = text("""
    SELECT user_id 
    FROM "User" 
    WHERE username = :username
""")

with engine.connect() as connection:
    result = connection.execute(query, {"username": username})
    row = result.fetchone()

if row is None:
    raise ValueError("Usu√°rio n√£o encontrado no banco.")

user_id = row[0]

print("User ID encontrado:", user_id)
```

```python
# Pegando tracks salvos (likes)
def get_all_liked_tracks(sp):
    tracks_data = []
    results = sp.current_user_saved_tracks(limit=50)
    
    while results:
        for item in results["items"]:
            track = item.get("track")
            if track is None:
                continue
            tracks_data.append({
                "track_name": track.get("name"),
                "artist": track.get("artists")[0].get("name") if track.get("artists") else None,
                "uri": track.get("uri"),
                "track_id": track.get("id")
            })
        if results.get("next"):
            results = sp.next(results)
        else:
            results = None
    return pd.DataFrame(tracks_data)

df_liked = get_all_liked_tracks(sp)

# Pegando playlists do usu√°rio que voc√™ realmente consegue acessar
accessible_playlists = []
results = sp.current_user_playlists(limit=50)
while results:
    for pl in results['items']:
        # s√≥ pega playlists do seu user_id (ou que voc√™ tem acesso)
        if pl['owner']['id'] == sp.current_user()['id']:
            accessible_playlists.append(pl)
    if results.get('next'):
        results = sp.next(results)
    else:
        results = None

# Pega todos os tracks dessas playlists
playlist_track_ids = set()
for pl in accessible_playlists:
    results = sp.playlist_items(pl['id'], fields="items.track.id,next", additional_types="track")
    tracks = results['items']
    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])
    for t in tracks:
        if t['track'] and t['track']['id']:
            playlist_track_ids.add(t['track']['id'])

# 4Ô∏è‚É£ Cria coluna booleano
df_liked['in_playlist'] = df_liked['track_id'].apply(lambda x: x in playlist_track_ids)

print(df_liked.head())
```
